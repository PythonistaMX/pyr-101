# Series de tiempo.

````r
Data <- read.csv("data/17/NASDAQ.csv")
````

````r
Data$Date <- as.POSIXct(as.Date(Data$Date, format = "%Y-%m-%d"))
````

````r
Serie <- Data[c(1,6)]
````

````r
Serie
````

## Wavelets.

````r
install.packages('wavelets')
````

````r
library(wavelets) 
````

````r
library(dplyr)
````

````r
install.packages("timelineR")
````

````r
require(timelineR)
````

````r
grafica <- plot_timeline(Serie)
````

````r
tsSerie = as.ts(Serie)
````

### La función ```dwt()```.

````r
help(dwt)
````

````r
model<- dwt (tsSerie, filter="la8", n.levels=5) 
````

````r
model
````

````r
plot(model)
````

https://en.wikipedia.org/wiki/Haar_wavelet

````r
model<- dwt (tsSerie, filter="haar", n.levels=7) 
````

````r
model
````

````r
plot(model)
````

````r
install.packages("seewave")
````

````r
library(seewave)
````

````r
model <- hilbert(tsSerie)
````

````r
summary(model)
````

````r
modeloReal <- Re(model)
modeloImag <- Im(model)
````
